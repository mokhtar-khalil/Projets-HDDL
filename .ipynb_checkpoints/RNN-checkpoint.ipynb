```python
def load_imdb_data(data_dir):
    """
    Charge les critiques et leurs labels à partir des fichiers texte.
    """
    reviews = []
    labels = []

    for label_type in ['pos', 'neg']:
        dir_name = os.path.join(data_dir, label_type)
        for fname in os.listdir(dir_name):
            if fname.endswith('.txt'):
                with open(os.path.join(dir_name, fname), encoding="utf-8") as f:
                    reviews.append(f.read())
                labels.append(1 if label_type == 'pos' else 0)

    return reviews, labels

# Chemin des dossiers train et test
train_dir = "/content/aclImdb/train"
test_dir = "/content/aclImdb/test"

# Charger les données
r_train, s_train = load_imdb_data(train_dir)
r_test, s_test = load_imdb_data(test_dir)
```

#### 4 - Supprimer les balises Html

```python
import re
def remove_html_tags(text):
    pattern = r'[^a-zA-Z0-9\s]'
    text = re.sub(pattern,'',text)
    return text
df['review'] = df['review'].apply(remove_html_tags)
dt['review'] = dt['review'].apply(remove_html_tags)
```

####  5 - Supprimer les URLs

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)

df['review'] = df['review'].apply(remove_url)
dt['review'] = dt['review'].apply(remove_url)
```

#### 6 - Supprimer les ponctuations

```python
import string,time
print(string.punctuation)
def remove_punctuation(text):
    return text.translate(str.maketrans('','',string.punctuation))
df['review'] = df['review'].apply(remove_punctuation)
dt['review'] = dt['review'].apply(remove_punctuation)
```

#### 7 - Supprimer les stopwords

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
# Convertir les stopwords en une chaîne de texte
stop_words_text = " ".join(stop_words)

# Générer le WordCloud à partir des stopwords
wordcloud = WordCloud().generate(stop_words_text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
```

```python
def stopwords_removal(text):
    new_text=[]
    for word in text.split():
        if word in stop_words:
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
df['review']=df['review'].apply(stopwords_removal)
dt['review']=dt['review'].apply(stopwords_removal)
df.head()
```

```python

# Define the model
RNN_model = Sequential()
RNN_model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=100))
RNN_model.add(SpatialDropout1D(0.2))
RNN_model.add(Bidirectional(SimpleRNN(64, return_sequences=True)))
RNN_model.add(Dropout(0.2))
RNN_model.add(BatchNormalization())
RNN_model.add(Bidirectional(SimpleRNN(32, return_sequences=True)))
RNN_model.add(Dropout(0.2))
RNN_model.add(BatchNormalization())
RNN_model.add(SimpleRNN(16, return_sequences=False))

# Continue with other layers
RNN_model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
RNN_model.add(Dense(1, activation='sigmoid'))

# Summarize the model
RNN_model.summary()
```

<span style="font-size:small; color:blue;">
    Cette architecture est une Bidirectional RNN conçue pour des données séquentielles, comme le texte. Elle utilise une couche d'Embedding pour encoder les mots en vecteurs, suivie de couches récurrentes bidirectionnelles (SimpleRNN) pour capturer les dépendances passées et futures dans les séquences. Des mécanismes de régularisation, comme le Dropout et la Batch Normalization, sont intégrés pour réduire le surapprentissage. Sa structure progressive (unités décroissantes) permet d'extraire des représentations de plus en plus abstraites.
    </span>

```python
RNN_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
start_time = time.time()
history_rnn = RNN_model.fit(X_train, y_train, epochs=15, batch_size=32, validation_data=(X_test, y_test))
end_time = time.time()
training_time_rnn = end_time - start_time
print(f"Temps total d'entraînement  rnn: {training_time_rnn:.2f} secondes")
```

```python
# Define the model
lstm_model = Sequential()
lstm_model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=100))
lstm_model.add(SpatialDropout1D(0.5))
lstm_model.add(LSTM(5, return_sequences=False))
lstm_model.add(Dropout(0.5))
lstm_model.add(BatchNormalization())


lstm_model.add(Dense(1, activation='sigmoid'))

# Summarize the model
lstm_model.summary()
```

<font size="2" color="blue">L'architecture LSTM utilise une couche d'Embedding pour transformer les mots en vecteurs numériques suivie d'une couche LSTM qui capture les relations temporelles dans les données séquentielles. La régularisation est assurée par le SpatialDropout1D et le Dropout, et la Batch Normalization stabilise l'entraînement. Enfin, une couche dense avec activation sigmoid permet une classification binaire.</font>

```python
lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
start_time = time.time()
history_lstm = lstm_model.fit(X_train, y_train, epochs=15, batch_size=128, validation_data=(X_test, y_test))

end_time = time.time()
training_time_lstm = end_time - start_time
print(f"Temps total d'entraînement  lstm: {training_time_lstm:.2f} secondes")
```

```python
# Define the model
GRU_model = Sequential()
GRU_model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=100))
GRU_model.add(SpatialDropout1D(0.5))
GRU_model.add(GRU(5, return_sequences=False))
GRU_model.add(Dropout(0.5))
#GRU_model.add(BatchNormalization())


GRU_model.add(Dense(1, activation='sigmoid'))

# Summarize the model
GRU_model.summary()
```

<span style="font-size: 12px; color: blue;"> L'architecture est basée sur une **GRU** (Gated Recurrent Unit), un type de réseau récurrent efficace pour modéliser les dépendances temporelles dans les séquences. Elle commence par une **couche d'Embedding** pour transformer les tokens en vecteurs denses, suivie d'une **couche GRU** pour capturer les relations dans les séquences. La régularisation est assurée par un **SpatialDropout** et un **Dropout** pour éviter le surapprentissage. Une couche Dense avec une activation sigmoid permet une classification binaire. Ce modèle est compact et performant pour des tâches de classification de texte.</span>

<div style="background-color: #e7f3fe; border: 1px solid #2196F3; padding: 15px; border-radius: 5px; box-shadow: 2px 2px 5px grey; text-align: center;">
    <b>Multi Layer Perceptron</b>
</div>


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau

early_stop = EarlyStopping(monitor='accuracy', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='loss', factor=0.5, patience=2, min_lr=1e-6)

MLP_model = Sequential()
MLP_model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=100))
MLP_model.add(Flatten())
MLP_model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))  # Régularisation L2
MLP_model.add(Dropout(0.5))  # Dropout renforcé
MLP_model.add(Dense(1, activation='sigmoid'))
# Compiler le modèle
MLP_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Résumé du modèle
MLP_model.summary()
start_time = time.time()
# Entraîner le modèle
history_mlp = MLP_model.fit(X_train, y_train, epochs=15, batch_size=36, validation_data=(X_test, y_test),callbacks = [early_stop,reduce_lr])

end_time = time.time()
training_time_mlp = end_time - start_time
print(f"Temps total d'entraînement  lstm: {training_time_mlp:.2f} secondes")
```

<font size="2" color ="blue" >L'architecture **MLP (Multi-Layer Perceptron)** est adaptée aux données vectorielles. Elle utilise une couche **Flatten** pour transformer les embeddings en vecteurs 1D, suivie d'une couche dense (256 neurones, activation **ReLU**) pour apprendre des représentations. La **Dropout (0.5)** limite le surapprentissage, et une sortie avec activation **sigmoïde** est utilisée pour la classification binaire.</font>

```python
min_train_acc = max(history_mlp.history['accuracy'])
min_val_acc = max(history_mlp.history['val_accuracy'])

# Plot training & validation accuracy values
plt.plot(history_mlp.history['accuracy'])
plt.plot(history_mlp.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(
    [f'Train accuracy MLP (min: {min_train_acc:.2f})', f'Validation accuracy MLP(max: {min_val_acc:.2f})'],
    loc='upper left'
)
plt.show()
```

<div style="background-color: #e7f3fe; border: 1px solid #2196F3; padding: 15px; border-radius: 5px; box-shadow: 2px 2px 5px grey; text-align: center;">
    <b>CNN</b>
</div>


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout

# Définir le modèle CNN
CNN_model = Sequential()
CNN_model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=100))
CNN_model.add(Conv1D(filters=128, kernel_size=5, activation='relu'))  # Convolutions 1D
CNN_model.add(GlobalMaxPooling1D())  # Pooling global
CNN_model.add(Dropout(0.5))  # Régularisation
CNN_model.add(Dense(1, activation='sigmoid'))  # Couche de sortie pour la classification binaire

# Compiler le modèle
CNN_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Résumé du modèle
CNN_model.summary()
start_time = time.time()


# Entraîner le modèle
history_cnn = CNN_model.fit(X_train, y_train, epochs=15, batch_size=256, validation_data=(X_test, y_test))

end_time = time.time()
training_time_cnn = end_time - start_time
print(f"Temps total d'entraînement  lstm: {training_time_cnn:.2f} secondes")
```

<span style="color:blue; font-size:small;">Cette architecture CNN est conçue pour traiter des données séquentielles telles que le texte. Elle utilise une **couche d'Embedding** pour encoder les mots, suivie d'une **couche de convolution 1D** pour extraire les motifs locaux dans les séquences. La couche **GlobalMaxPooling1D** réduit la dimensionnalité tout en capturant les caractéristiques les plus importantes, et le **Dropout** agit comme régularisateur pour éviter le surapprentissage. Enfin, une couche de sortie sigmoïde réalise la classification binaire.</span>


```python
max_train_acc = max(history_cnn.history['accuracy'])
max_val_acc = max(history_cnn.history['val_accuracy'])
# Plot training & validation accuracy values
plt.plot(history_cnn.history['accuracy'])
plt.plot(history_cnn.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(
    [f'Train accuracy CNN (max: {max_train_acc:.2f})', f'Validation accuracy CNN(max: {max_val_acc:.2f})'],
    loc='upper left'
)
plt.show()
```

```python
min_train_loss = min(history_cnn.history['loss'])
min_val_loss = min(history_cnn.history['val_loss'])

# Plot training & validation accuracy values
plt.plot(history_cnn.history['loss'])
plt.plot(history_cnn.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('Epoch')
# Ajouter les valeurs minimales dans la légende
plt.legend(
    [f'Train loss cnn (min: {min_train_loss:.2f})', f'Validation loss cnn(min: {min_val_loss:.2f})'],
    loc='upper left'
)
plt.xlim(0,10)
plt.ylim(0,1)
plt.show()
```

   Ensemble vote classifieur par vote majoritaire
</div>

```python
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np

class KerasClassifierWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, model, batch_size=32, epochs=10, verbose=0):
        self.model = model
        self.batch_size = batch_size
        self.epochs = epochs
        self.verbose = verbose

    def fit(self, X, y):
        # Entraînez le modèle Keras
        self.model.fit(X, y, batch_size=self.batch_size, epochs=self.epochs, verbose=self.verbose)
        return self

    def predict(self, X):
        # Retourne les classes binaires (0 ou 1)
        return (self.model.predict(X) > 0.5).astype(int).flatten()

    def predict_proba(self, X):
        # Retourne les probabilités prédictes
        return self.model.predict(X)
```

```python
from mlxtend.classifier import EnsembleVoteClassifier
from sklearn.metrics import accuracy_score

# Encapsulation des modèles Keras
cnn_clf = KerasClassifierWrapper(CNN_model, epochs=5)
rnn_clf = KerasClassifierWrapper(RNN_model, epochs=5)
gru_clf = KerasClassifierWrapper(GRU_model, epochs=5)
mlp_clf = KerasClassifierWrapper(MLP_model, epochs=5)
lstm_clf = KerasClassifierWrapper(lstm_model, epochs=5)

# Créer l'ensemble avec vote majoritaire
eclf = EnsembleVoteClassifier(clfs=[cnn_clf, rnn_clf, gru_clf,mlp_clf,lstm_clf], weights=[1, 1, 1,1,1], fit_base_estimators=True)

# Entraînez l'ensemble
eclf.fit(X_train, y_train)

# Prédisez sur les données de test
y_pred = eclf.predict(X_test)

# Évaluez l'accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy de l'ensemble : {accuracy:.2f}")
```
